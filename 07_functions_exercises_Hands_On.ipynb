## Exercise 1 <--- Homeowrk 4/7/2024

Write a function called `is_common(s1, s2)`, that gets two strings `s1` and `s2` and returns whether `s1` and `s2` have a common letters, or not. The test should be case insensitive (i.e., "A" and "a" should be counted the same).

````python
# My Solution

def is_common(s1, s2):
    common_letters = []
    for char in s1.lower():
        if char in s2.lower() and char not in common_letters:
            common_letters.append(char)
    if common_letters:
        return True, common_letters
    else:
        return False

is_common('international', 'nation')
````

## Exercise 3 <--- Homeowrk 4/7/2024

Write a function that receives an integer `n` and returns the number of primes up to `n`.
```
n = 10

    2
    3
    5
    7
    
--> 4
```

````python
def is_prime_num(n):
  if n == 1:
    return False
  for d in range(2, int(n**0.5)+1):
    if n % d == 0:
      return False
  return True
````

````python
def count_prime(n):
  counter = 0
  for x in range(1, n+1):
    if is_prime_num(x):
      counter = counter + 1
    return counter
````

````python
count_prime(7)
````

### Solution

````python
def is_prime(n):

    # The number 1 is not a prime number.
    if n == 1:
        return False

    for d in range(2, int(n**0.5)+1):
        if n % d == 0:
            return False

    return True
````

````python
def count_primes(n):

    counter = 0

    for x in range(1, n+1):

        if is_prime(x):
            counter = counter + 1

    return counter
````

## Exercise 2 ---> Homework 9/7/2024

Write a function *is_sum_of_digits_even* which sums up the digits of a number and return *True* if the sum is an even number, or *False* if the sum is an odd number.
```
123 --> 1 + 2 + 3 = 6 --> True
45 --> 4 + 5 = 9 --> False
```

````python
"My Solution"

def is_sum_of_digits_even(num):
  str_num = str(num)
  sum_of_digits = 0
  for char in str_num:
    sum_of_digits += float(char)
  if  sum_of_digits % 2 == 0:
    return num, sum_of_digits, True
  else:
    return num, sum_of_digits, False

is_sum_of_digits_even(427)
````

### Solution

````python
def is_sum_of_digits_event(num):

    sum_of_digits = 0
    for digit_char in str(num):
        sum_of_digits += int(digit_char)

    return sum_of_digits % 2 == 0
````
